```python
import pandas as pd
from time import time
```

```python
pd.__version__
```

```python
df = pd.read_csv('yellow_tripdata_2021-01.csv/yellow_tripdata_2021-01.csv', nrows=100)
```

```python
df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime)
df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)
```

```python
df.head()
```

```python
print(pd.io.sql.get_schema(df, 'yellow_taxi_data'))
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
engine.connect()
```

```python
connection = engine.raw_connection()
```

```python
query = """
SELECT count(1) FROM yellow_taxi_trips;
"""

pd.read_sql(query, con=connection)
```

```python
print(pd.io.sql.get_schema(df, name='yellow_taxi_data', con=engine))
```

```python
# chunk dataframe to insert it chunk by chunk not one million at a time
df_iter = pd.read_csv('yellow_tripdata_2021-01.csv/yellow_tripdata_2021-01.csv', iterator=True, chunksize=100000)
```

```python
df_iter
```

```python
df = next(df_iter)
```

```python
df
```

```python
df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime)
df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)
```

```python
print(pd.io.sql.get_schema(df, name='yellow_taxi_data', con=engine))
```

```python
df.head(n=0).to_sql(name='yellow_taxi_data', con=engine, if_exists='replace')
```

```python
%time df.to_sql(name='yellow_taxi_data', con=engine, if_exists='append')
```

```python
while True: 
    t_start = time()

    df = next(df_iter)

    df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)
    df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime)
    
    df.to_sql(name='yellow_taxi_data', con=engine, if_exists='append')

    t_end = time()

    print('inserted another chunk, took %.3f second' % (t_end - t_start))
```

```python
query = """
SELECT *
FROM pg_catalog.pg_tables
WHERE schemaname != 'pg_catalog' AND 
    schemaname != 'information_schema';
"""

pd.read_sql(query, con=connection)
```
